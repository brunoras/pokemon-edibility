```python
import pandas as pd
import re
from collections import OrderedDict
```

```python
entradas = pd.read_csv('bulbapedia.csv')
entradas.head()
```

```python
def remover(textos, regex):
    if type(textos) == str:
        return regex.sub("", textos)
    else:
        return [regex.sub("", texto) for texto in textos]
```

```python
regex_quebra = re.compile(r"(\n)")
lista = remover(lista, regex_quebra)
lista
```

```python
pkmn_dict = OrderedDict()
lista_1 = []
regex_quebra = re.compile(r"(\n)")
 
for i in range(len(entradas)):
   lista_1 = entradas.iloc[i,:].dropna().unique().tolist()
   lista_1 = remover(lista_1, regex_quebra)
   pkmn_dict[lista_1[0]] = lista_1[1:]
pkmn_dict
```

```python
html_keys = {"Nidoran%E2%99%80" : "Nidoran♀",
"Nidoran%E2%99%82" : "Nidoran♂", 
"Farfetch%27d" : "Farfetch'd", 
"Mr._Mime" : "Mr. Mime",
"Mime_Jr." : "Mime Jr.",
"Flab%C3%A9b%C3%A9" : "Flabébé",
"Type:_Null" : "Type: Null",
"Tapu_Koko" : "Tapu Koko",
"Tapu_Lele" : "Tapu Lele",
"Tapu_Bulu" : "Tapu Bulu",
"Tapu_Fini" : "Tapu Fini",
"Sirfetch%27d" : "Sirfetch'd",
"Mr._Rime" : "Mr. Rime"}
 
html_keys.keys()
 
for key in html_keys.keys():
   pkmn_dict[html_keys.get(key)] = pkmn_dict.pop(key)
pkmn_dict
```

```python
bulba_frame = pd.DataFrame.from_dict(pkmn_dict, orient='index')
bulba_frame.head()
```

```python
compression_opts = dict(method='zip', archive_name='new_bulbapedia.csv')  
bulba_frame.to_csv('new_bulbapedia.zip', index=True, compression=compression_opts)
```
